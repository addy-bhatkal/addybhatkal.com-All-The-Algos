```python
import pandas as pd, numpy as np, os, warnings, psutil
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import re
import pickle
from sklearn.svm import SVC
```

```python
df = pd.read_csv('/Users/adityabhatkal/Desktop/Practice 32/AG_news_classification.csv')
```

```python
df['Class Index'] = df['Class Index'].map({3: 'Finance', 2: 'Sports', 1:'Geopolitics', 4:'Tech' })
```

```python
df
```

```python
df['Query'] = df['Title'] + " " + df['Description']
```

```python
df1 = pd.DataFrame([df['Class Index'], df['Query']]).transpose()
```

```python
df1
```

```python
df1['Query'] = df1['Query'].apply(lambda x : re.sub(r'\d+', '', x))
```

```python
cv1 = CountVectorizer(ngram_range=(1,1), stop_words='english', max_features=800)
vec_toarray = cv1.fit_transform(df1.iloc[:,1]).toarray()
matrix = pd.DataFrame(vec_toarray, columns=[cv1.get_feature_names_out()])
matrix
```

```python
df1
```

```python
x = matrix
y = df1['Class Index']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
nb = MultinomialNB()
model = nb.fit(x_train, y_train)
y_predict = nb.predict(x_test)
acc = accuracy_score(y_predict, y_test)
acc
```

```python
def sentiment():
    user_input = input()
    new_user_row = pd.DataFrame({'Class Index':[''], 'Query': [user_input]})
    df2 = pd.concat([df1,new_user_row], ignore_index = True)

    cv1 = CountVectorizer(ngram_range=(1,1), stop_words='english', max_features=800)
    vec_toarray = cv1.fit_transform(df2.iloc[:,1]).toarray()
    matrix = pd.DataFrame(vec_toarray, columns=[cv1.get_feature_names_out()])
    matrix

    x = matrix
    y = df2['Class Index']
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
    nb = MultinomialNB()
    model = nb.fit(x_train, y_train)
    y_predict = nb.predict(x_test)
    acc = accuracy_score(y_predict, y_test)
    acc

    vec_toarray = cv1.transform([user_input])
    return model.predict(vec_toarray.toarray())[0]
```

```python
sentiment()
```

```python
with open('nb_model.pkl', 'wb') as model_file, open('cv1.pkl', 'wb') as vec_file:
    pickle.dump(cv1 ,vec_file)
    pickle.dump(model ,model_file)


with open('nb_model.pkl', 'rb') as model_file, open('cv1.pkl', 'rb') as vec_file:
    cv2 = pickle.load(vec_file)
    model2 = pickle.load(model_file)
```

```python
def senti():
    user_input1 = input()
    vec1 = cv2.transform([user_input1]).toarray()
    user_row = model2.predict(vec1)
    return user_row[0]
```
